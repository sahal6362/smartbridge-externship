```python
import pandas
import matplotlib
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
df=sns.load_dataset('iris')
df.head()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
plt.scatter(df.sepal_length,df.petal_length)
sns.set_style('darkgrid')
```

```python
plt.scatter(df.sepal_length,df.petal_length)
sns.set_style('whitegrid')
plt.show()
```

```python
from matplotlib import pyplot
import seaborn as sns
plt.scatter(df.sepal_length,df.petal_length)
sns.set_style('darkgrid',{'grid.color':'.1'})
plt.show()
```

```python
sns.relplot(data=df,x="sepal_length",y='petal_length')
```

```python
sns.relplot(data=df,x="sepal_length",y='petal_length',hue='species')
```

```python
df.species.unique()
```

```python
sns.kdeplot(df.loc[(df['species']=='setosa'),'sepal_length'],color='b',Label='Setosa')
```

```python
import matplotlib.pyplot as plt
sns.set_style('dark')
sns.kdeplot(df.loc[(df['species']=='setosa'),'sepal_length'],color='b',Label='Setosa')
sns.kdeplot(df.loc[(df['species']=='virginica'),'sepal_length'],shade=True,color='b',Label='Verginica')
```

```python
import matplotlib.pyplot as plt
sns.set_style('dark')
sns.kdeplot(df.loc[(df['species']=='setosa'),'sepal_length'],shade=True,color='b',Label='Setosa')
sns.kdeplot(df.loc[(df['species']=='virginica'),'sepal_length'],shade=True,color='b',Label='Verginica')
```

```python
sns.boxplot(x='sepal_length',y='species',data=df)
```
